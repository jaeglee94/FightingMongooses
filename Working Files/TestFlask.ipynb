```python
from flask import (
    Flask,
    render_template,
    jsonify,
    request,
    redirect)
```

```python
import sqlite3
```

```python
app = Flask(__name__)
```

```python
app.route("/")
def home():
    return render_template("index.html")
```

```python
#connect
conn = sqlite3.connect("routes.db")
```

```python
#obtain a cursor - something to loop through via database connection
cur = conn.cursor()
```

```python
cur.execute("SELECT name FROM sqlite_master WHERE type='table'")
#obtain data from cursor via loop - in this case - table names in database file
rows = cur.fetchall()
 
for row in rows:
    print(row)
```

```python
#obtain details of the schema/table format from table named player
cur.execute("SELECT sql FROM sqlite_master WHERE name='Routes'")
headers = cur.fetchall()
 
for item in headers:
    print(item)
```

```python
mytext = headers[0][0]

headers = mytext.replace("CREATE TABLE Routes(\n  \"", "").replace("\" TEXT", "").replace("\n  \"","").replace("\n)","").split(",")

```

```python
headers
```

```python
#print all results
cur.execute("SELECT * FROM Routes")
rows = cur.fetchall()
```

```python
data = dict()
count = 0

for item in headers:
    columns = []
    for row in rows:
        columns.append(row[count])
    data[item] = columns
    count += 1

data
```

```python
# create route that returns data for plotting
@app.route("/api/")
def final_data():

    return jsonify(data)


if __name__ == "__main__":
    app.run()
```
